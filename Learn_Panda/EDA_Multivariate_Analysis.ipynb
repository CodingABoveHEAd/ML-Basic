```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")

tips
```

```python
#Bivariate Analysis
# Scatter Plot(Numerical vs Numerical)
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex",style="smoker")
#here we are using hue and style to add more dimensions to the scatter plot(multivariate analysis)
```

```python
titanic.head()
```

```python
#Numerical vs Categorical
#x-axis -> Categorical
#y-axis -> Numerical
sns.barplot(data=titanic, x="pclass", y="age",hue="sex")
```

```python
#categorical vs numerical
#boxplot
sns.boxplot(data=titanic, x="sex", y="age",)
```

```python
#distplot categorical vs numerical
sns.distplot(titanic[titanic['survived']==0]['age'],label='Not Survived',hist=False)
sns.distplot(titanic[titanic['survived']==1]['age'],label='Survived',hist=False)
#blue line -> not survived
#orange line -> survived
```

```python
#categorical vs categorical
#x-axis -> Categorical
#y-axis -> Categorical

titanic.groupby(['pclass','sex'])['survived'].mean()
```

```python
#numerical vs numerical vs categorical
sns.pairplot(iris, hue="species")
```
