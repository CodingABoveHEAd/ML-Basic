```python
import pandas as pd
import seaborn as sns
csv1=pd.read_csv('tested.csv')
csv1
```

```python
csv1['Survived'].value_counts()
sns.countplot(csv1, x="Sex")
```

```python
csv1['Pclass'].value_counts().plot(kind='pie', autopct='%.2f%%')
#Thats it for categorical data
```

```python
#For numerical data
#Histogram
#tells the exact distribution of a numerical column
import matplotlib.pyplot as plt
plt.hist(csv1['Age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
```

```python
#Probality Density Function
sns.distplot(csv1['Age'])
plt.xlabel('Age')
plt.ylabel('Probability Density')
plt.title('Distribution of Age')
```

```python
sns.boxplot(x=csv1['Age'])
plt.xlabel('Age')
plt.title('Boxplot of Age')
```
