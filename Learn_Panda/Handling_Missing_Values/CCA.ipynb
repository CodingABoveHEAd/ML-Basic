```python
import pandas as pd
import numpy as np

df=pd.read_csv('data_science_job.csv')
df
```

```python
df.isnull().mean()*100
```

```python
df.shape
df.columns
```

```python
col=[f 
     for f in df.columns 
     if df[f].isnull().mean()<0.05 and df[f].isnull().mean()>0]

df1=df[col]
df1.isnull().mean()*100
```

```python
#after drop 89% data will remain 
len(df1.dropna())/len(df)
```

```python
new_df=df1.dropna()
df1.shape,new_df.shape
```

```python


import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)

df['training_hours'].hist(bins=30, ax=ax, density=True, color='red')
new_df['training_hours'].hist(bins=30, ax=ax, density=True)

plt.show()
```
