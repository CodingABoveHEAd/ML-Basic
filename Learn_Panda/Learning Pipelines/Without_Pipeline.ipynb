```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('tested.csv')
df.shape
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns=['Survived']),
    df['Survived'],
    test_size=0.2,
    random_state=42
)
```

```python
df.isnull().sum()
```

```python
si_age=SimpleImputer(strategy='mean')
si_Fare=SimpleImputer(strategy='mean')

X_train_age=si_age.fit_transform(X_train[['Age']])
X_train_fare=si_Fare.fit_transform(X_train[['Fare']])

X_test_age=si_age.transform(X_test[['Age']])
X_test_fare=si_Fare.transform(X_test[['Fare']])

# X_train[['Age']]=X_train_age
# X_train[['Fare']]=X_train_fare

# X_test[['Age']]=X_test_age
# X_test[['Fare']]=X_test_fare
```

```python
ohe_sex=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
X_train_sex=ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked=ohe_embarked.fit_transform(X_train[['Embarked']])

X_test_sex=ohe_sex.transform(X_test[['Sex']])
X_test_embarked=ohe_embarked.transform(X_test[['Embarked']])

```

```python
X_train_remain=X_train.drop(columns=['Age','Sex','Embarked'])
X_test_remain=X_test.drop(columns=['Age','Sex','Embarked'])

X_train
```

```python
X_train_transformed=np.concatenate(
    [X_train_remain,X_train_age,X_train_fare,X_train_sex,X_train_embarked],axis=1
)

X_test_transformed=np.concatenate(
    [X_test_remain,X_test_age,X_test_fare,X_test_sex,X_test_embarked],axis=1
)
```

```python

from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)

y_pred=clf.predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
