```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df=pd.read_csv('customer.csv')
df.head()
```

```python
df=df.iloc[:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('purchased', axis=1), df['purchased'], test_size=0.2, random_state=0)
X_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

# specify order of categories
ordinal_encoder = OrdinalEncoder(
    categories=[['Poor','Average','Good'], ['School','UG','PG']]
)

# select only categorical columns
cat_cols = ['review', 'education']

# fit only on categorical columns
ordinal_encoder.fit(X_train[cat_cols])

# transform only categorical columns
X_train_enc = ordinal_encoder.transform(X_train[cat_cols])
X_test_enc = ordinal_encoder.transform(X_test[cat_cols])

X_train_enc=pd.DataFrame(X_train_enc, columns=cat_cols)
X_test_enc=pd.DataFrame(X_test_enc, columns=cat_cols)

X_train_enc.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(y_train)
y_train_enc=le.transform(y_train)
y_test_enc=le.transform(y_test)

y_train_enc
```
