```python
import pandas as pd
import requests

response = requests.get('https://api.themoviedb.org/3/movie/top_rated?language=en-US&page=1&api_key=5dca48129385e332cb9d1ed29f7c78ef')

pd.DataFrame(response.json()['results'])[[ 'id', 'title', 'release_date', 'video', 'popularity', 'vote_average' ]]
```

```python
df1=pd.DataFrame()
for i in range(1,495):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?language=en-US&page={i}&api_key=5dca48129385e332cb9d1ed29f7c78ef')
    page_data=pd.DataFrame(response.json()['results'])[[ 'id', 'title', 'release_date', 'video', 'popularity', 'vote_average' ]]
    df1 = pd.concat([df1, page_data], ignore_index=True)

df1.to_csv('top_rated_movies.csv', index=False)
```
