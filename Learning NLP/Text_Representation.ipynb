```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({'Text':['people watch youtube','youtube watch youtube',
'people write comments','youtube write comments'],'output':[1,1,0,0]})
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer(ngram_range=(3,3))
#bag of words
bow=vect.fit_transform(df['Text'])
```

```python
print(vect.vocabulary_)
sorted(vect.vocabulary_.items())
```

```python
print(bow.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect=TfidfVectorizer()
tfidf=tfidf_vect.fit_transform(df['Text'])
tfidf_vect.fit_transform(df['Text']).toarray()
#tf=frequency of word in document/ total words in document
#idf=log(total documents/ documents with word in it) + 1
```
